## Application 1: Multi-Agent Marketplace

A real-world scenario with multiple buyer and seller agents.

```python
from typing import Dict, List
from dataclasses import dataclass

@dataclass
class Listing:
    listing_id: str
    seller_agent_id: str
    product: str
    price: float
    quantity: int

@dataclass
class Transaction:
    buyer_agent_id: str
    seller_agent_id: str
    product: str
    price: float
    timestamp: object = None

class Marketplace:
    """Real-world application: Multi-agent marketplace with infrastructure."""
    
    def __init__(self):
        self.listings: Dict[str, Listing] = {}
        self.transactions: List[Transaction] = []
        self.price_history: Dict[str, List[float]] = {}  # product -> prices
        self.agent_ratings: Dict[str, float] = {}
    
    def list_product(self, seller_id: str, product: str, 
                    price: float, quantity: int) -> str:
        """Agent posts product for sale."""
        # Infrastructure checks: seller has permission
        listing_id = f"listing_{len(self.listings)}"
        listing = Listing(listing_id, seller_id, product, price, quantity)
        self.listings[listing_id] = listing
        
        # Track price history
        if product not in self.price_history:
            self.price_history[product] = []
        self.price_history[product].append(price)
        
        print(f"✓ Listed: {product} @ ${price} (seller: {seller_id})")
        return listing_id
    
    def execute_transaction(self, buyer_id: str, seller_id: str,
                           product: str, price: float) -> bool:
        """Execute a transaction between agents."""
        # Infrastructure checks
        if not self._validate_transaction(buyer_id, seller_id, price):
            return False
        
        txn = Transaction(buyer_id, seller_id, product, price)
        self.transactions.append(txn)
        print(f"✓ Sold: {product} from {seller_id} to {buyer_id} @ ${price}")
        return True
    
    def _validate_transaction(self, buyer_id: str, seller_id: str,
                             price: float) -> bool:
        """Infrastructure validation."""
        # Check for price gouging
        if product := self._get_product_for_price(price):
            avg_price = sum(self.price_history[product]) / len(self.price_history[product])
            if price > avg_price * 2:
                print(f"  ⚠ Price gouging detected: ${price} vs avg ${avg_price}")
                return False
        
        return True
    
    def _get_product_for_price(self, price: float):
        for product, prices in self.price_history.items():
            if price in prices:
                return product
        return None
    
    def get_marketplace_stats(self):
        print("\n=== Marketplace Statistics ===")
        print(f"Listings: {len(self.listings)}")
        print(f"Transactions: {len(self.transactions)}")
        print(f"Total volume: ${sum(t.price for t in self.transactions):.2f}")

# Test marketplace
marketplace = Marketplace()

# Seller agents list products
marketplace.list_product("seller_A", "laptop", 999, 5)
marketplace.list_product("seller_B", "laptop", 995, 3)
marketplace.list_product("seller_A", "mouse", 25, 20)

# Buyer agents execute transactions
marketplace.execute_transaction("buyer_1", "seller_A", "laptop", 999)
marketplace.execute_transaction("buyer_2", "seller_B", "laptop", 995)
marketplace.execute_transaction("buyer_3", "seller_A", "laptop", 2000)  # Will be blocked
marketplace.execute_transaction("buyer_1", "seller_A", "mouse", 25)

marketplace.get_marketplace_stats()
```

## Application 2: Governance Dashboard

Human operators monitor and control agent systems.

```python
class GovernanceDashboard:
    """Real-world application: Oversight and control."""
    
    def __init__(self):
        self.agent_status: Dict[str, str] = {}  # running, paused, blocked
        self.interventions: List[Dict[str, str]] = []
        self.compliance_reports: List[Dict[str, any]] = []
    
    def register_agent(self, agent_id: str):
        """Register agent for monitoring."""
        self.agent_status[agent_id] = "running"
        print(f"✓ Monitoring {agent_id}")
    
    def pause_agent(self, agent_id: str, reason: str):
        """Pause an agent."""
        if agent_id in self.agent_status:
            self.agent_status[agent_id] = "paused"
            self.interventions.append({
                "agent": agent_id,
                "action": "pause",
                "reason": reason,
                "timestamp": str(datetime.now())
            })
            print(f"✓ Paused {agent_id}: {reason}")
    
    def block_agent(self, agent_id: str, reason: str):
        """Block an agent from taking actions."""
        if agent_id in self.agent_status:
            self.agent_status[agent_id] = "blocked"
            self.interventions.append({
                "agent": agent_id,
                "action": "block",
                "reason": reason,
                "timestamp": str(datetime.now())
            })
            print(f"✓ Blocked {agent_id}: {reason}")
    
    def print_dashboard(self):
        print("\n=== Governance Dashboard ===")
        print("\nAgent Status:")
        for agent_id, status in self.agent_status.items():
            symbol = "🟢" if status == "running" else "🟡" if status == "paused" else "🔴"
            print(f"  {symbol} {agent_id}: {status}")
        
        if self.interventions:
            print(f"\nRecent Interventions ({len(self.interventions)}):")
            for intervention in self.interventions[-3:]:
                print(f"  - {intervention['agent']}: {intervention['action']} ({intervention['reason']})")

# Test governance
dashboard = GovernanceDashboard()

dashboard.register_agent("agent_1")
dashboard.register_agent("agent_2")
dashboard.register_agent("agent_3")

# Detect anomaly and pause agent
dashboard.pause_agent("agent_2", "Unusual activity detected")
dashboard.block_agent("agent_3", "Policy violation")

dashboard.print_dashboard()
```

## Application 3: Production Deployment Pattern

How to deploy agents with proper infrastructure.

```python
class ProductionAgent:
    """Agent deployed with full infrastructure support."""
    
    def __init__(self, agent_id: str, infrastructure: Dict[str, any]):
        self.agent_id = agent_id
        self.registry = infrastructure.get("registry")
        self.audit = infrastructure.get("audit")
        self.permissions = infrastructure.get("permissions")
        self.monitor = infrastructure.get("monitor")
    
    def perform_action(self, action_type: str, params: Dict[str, any]) -> bool:
        """Take an action with infrastructure support."""
        
        # 1. CHECK PERMISSIONS
        if not self.permissions.has_permission(self.agent_id, action_type):
            print(f"  ✗ Permission denied: {action_type}")
            return False
        
        # 2. CHECK LIMITS
        for param, value in params.items():
            if not self.permissions.check_limit(self.agent_id, param, value):
                print(f"  ✗ Limit exceeded: {param}={value}")
                return False
        
        # 3. PERFORM ACTION
        result = f"Executed {action_type} with {params}"
        print(f"  ✓ Action performed: {action_type}")
        
        # 4. LOG ACTION
        self.audit.log_action(self.agent_id, action_type, params, result)
        
        # 5. MONITOR
        self.monitor.record_action(self.agent_id, action_type)
        
        return True

print("\n=== Production Deployment Pattern ===")
print("\nComponent 1: Security (Permissions & Limits)")
print("Component 2: Logging (Audit Trail)")
print("Component 3: Monitoring (Behavior Tracking)")
print("Component 4: Governance (Human Control)")
print("\nThis ensures agents operate safely at scale.")
```
