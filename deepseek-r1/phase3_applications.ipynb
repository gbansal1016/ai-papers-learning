## Application 1: Mathematical Reasoning

Solving word problems with step-by-step reasoning.

### IMPORTANT: This is a Procedural Simulation

**No actual machine learning model is used here.** The `evaluate()` method is entirely procedural—it uses the `model_skill` parameter as input and performs pure mathematical calculations with random noise to simulate outcomes. Think of it as a mock evaluation function that demonstrates:

- How reward signals work in GRPO
- How reasoning depth affects performance
- How to structure domain-specific evaluation functions

In a real system, the `evaluate()` function would call an actual language model (like DeepSeek) and check its actual output against correct answers. Here, we simulate that behavior procedurally.

```python
class MathProblemSolver:
    """Simulate solving math problems with varying difficulty.
    
    NOTE: This is a PROCEDURAL SIMULATION. No actual model is used.
    The evaluate() method uses mathematical formulas and random noise
    to simulate how a model with skill level X would perform.
    
    In production, you would replace evaluate() with calls to a real
    language model and actual answer verification.
    """
    
    def __init__(self):
        self.problems = [
            {
                'id': 1,
                'question': 'What is 7 × 8?',
                'answer': 56,
                'difficulty': 0.1,
                'min_steps': 1,
                'category': 'arithmetic'
            },
            {
                'id': 2,
                'question': 'Sarah has 12 apples. She gives half to Tom. Tom gives a third to Lisa. How many does Lisa have?',
                'answer': 2,
                'difficulty': 0.3,
                'min_steps': 3,
                'category': 'word_problem'
            },
            {
                'id': 3,
                'question': 'If a train travels 60 km/h for 2.5 hours, then 80 km/h for 1.5 hours, what is the average speed?',
                'answer': 70,
                'difficulty': 0.5,
                'min_steps': 4,
                'category': 'physics'
            },
        ]
    
    def evaluate(
        self,
        problem_id: int,
        reasoning_steps: int,
        model_skill: float = 0.7
    ) -> Dict[str, Any]:
        """Evaluate a math problem solution.
        
        This is PURELY PROCEDURAL - it does NOT call any model.
        It takes problem parameters and model_skill, then simulates
        what the outcome would be using math + random noise.
        
        Args:
            problem_id: which problem to evaluate
            reasoning_steps: how many steps the model took
            model_skill: underlying capability (0-1 scale)
        
        Returns:
            dict with 'correct', 'reward', 'efficiency', 'problem'
        """
        problem = self.problems[problem_id % len(self.problems)]
        difficulty = problem['difficulty']
        min_steps = problem['min_steps']
        
        # Simulate: does model have enough reasoning steps?
        has_enough_steps = reasoning_steps >= max(1, min_steps - 1)
        
        # Simulate: base accuracy from skill + step bonus/penalty
        base_accuracy = model_skill + (0.3 if has_enough_steps else -0.2)
        base_accuracy = np.clip(base_accuracy, 0, 1)
        
        # Simulate: add noise to make outcomes realistic/variable
        actual_accuracy = base_accuracy + np.random.randn() * 0.1
        actual_accuracy = np.clip(actual_accuracy, 0, 1)
        
        # Simulate: is the answer correct?
        is_correct = actual_accuracy > 0.5
        
        # Efficiency bonus: correct with minimal steps preferred
        efficiency = 1.0 / (1 + reasoning_steps / 10)
        
        reward = 1.0 if is_correct else 0.0
        
        return {
            'correct': is_correct,
            'reward': reward,
            'efficiency': efficiency,
            'problem': problem['question'][:50]
        }

# Demonstrate
math_solver = MathProblemSolver()

print("Mathematical Reasoning Task (Procedural Simulation)\n" + "="*50)
for problem in math_solver.problems:
    results = {
        'quick': math_solver.evaluate(problem['id'], 1, model_skill=0.7),
        'thorough': math_solver.evaluate(problem['id'], 5, model_skill=0.7)
    }
    
    print(f"\nProblem {problem['id']}: {problem['question'][:40]}...")
    print(f"  Quick (1 step):     Correct: {results['quick']['correct']}, Reward: {results['quick']['reward']}")
    print(f"  Thorough (5 steps): Correct: {results['thorough']['correct']}, Reward: {results['thorough']['reward']}")

print("\n⚠️  Note: These results are SIMULATED. No actual model was called.")
```

## Application 2: Logical Reasoning

Multi-step logic puzzles.

### NOTE: Procedural Simulation
Like Application 1, this is a **procedural simulation**. The `evaluate()` method does not call any real model—it uses mathematical formulas to simulate how a model with varying skill levels would perform on logic puzzles.

```python
class LogicPuzzleSolver:
    """Logic puzzle solving with GRPO.
    
    NOTE: Like Application 1, this is also a PROCEDURAL SIMULATION.
    """
    
    def __init__(self):
        self.puzzles = [
            {
                'id': 1,
                'name': 'Simple Deduction',
                'difficulty': 0.3,
                'min_steps': 2
            },
            {
                'id': 2,
                'name': 'Multi-step Logic',
                'difficulty': 0.6,
                'min_steps': 4
            },
            {
                'id': 3,
                'name': 'Complex Reasoning',
                'difficulty': 0.8,
                'min_steps': 6
            },
        ]
    
    def evaluate(self, puzzle_id: int, reasoning_steps: int, model_skill: float = 0.7) -> bool:
        """Solve a logic puzzle (PROCEDURAL SIMULATION)."""
        puzzle = self.puzzles[puzzle_id % len(self.puzzles)]
        
        # Difficulty depends on having enough steps
        base_score = model_skill - puzzle['difficulty']
        step_bonus = min((reasoning_steps / puzzle['min_steps']) * 0.5, 0.5)
        
        final_score = base_score + step_bonus + np.random.randn() * 0.15
        final_score = np.clip(final_score, 0, 1)
        
        return final_score > 0.5

# Demonstrate
logic_solver = LogicPuzzleSolver()

print("\nLogic Reasoning Tasks (Procedural Simulation)\n" + "="*50)
for puzzle in logic_solver.puzzles:
    success_rates = {}
    for steps in [1, 3, 5, 10]:
        successes = sum([
            logic_solver.evaluate(puzzle['id'], steps, model_skill=0.7)
            for _ in range(20)
        ])
        success_rates[steps] = successes / 20
    
    print(f"\n{puzzle['name']} (min {puzzle['min_steps']} steps):")
    for steps, rate in success_rates.items():
        print(f"  {steps:2d} steps: {rate:.0%}")
```

## Application 3: Code Generation

Reasoning about algorithms and implementations.

### NOTE: Procedural Simulation
This is also a **procedural simulation**. No code is actually generated or executed. The `evaluate()` method simulates correctness and efficiency scores based on the model's skill and reasoning depth.

```python
class CodeGenerationTask:
    """Code generation with verification (PROCEDURAL SIMULATION)."""
    
    def __init__(self):
        self.tasks = [
            {
                'id': 1,
                'name': 'Sum of Array',
                'complexity': 0.1,
                'min_steps': 2
            },
            {
                'id': 2,
                'name': 'Fibonacci Generator',
                'complexity': 0.4,
                'min_steps': 4
            },
            {
                'id': 3,
                'name': 'Dynamic Programming',
                'complexity': 0.7,
                'min_steps': 6
            },
        ]
    
    def evaluate(self, task_id: int, reasoning_steps: int, model_skill: float = 0.7) -> Dict:
        """Evaluate code generation (PROCEDURAL SIMULATION)."""
        task = self.tasks[task_id % len(self.tasks)]
        
        # Probability of correct code
        base_prob = model_skill - task['complexity']
        reasoning_help = min((reasoning_steps / task['min_steps']) * 0.3, 0.3)
        
        correctness = np.clip(base_prob + reasoning_help + np.random.randn() * 0.15, 0, 1)
        
        # Efficiency: fewer lines is better
        efficiency_score = max(0, 1 - (reasoning_steps / 10))
        
        return {
            'correct': correctness > 0.5,
            'correctness_score': correctness,
            'efficiency_score': efficiency_score,
            'tests_passed': int(correctness * 10)
        }

# Demonstrate
code_gen = CodeGenerationTask()

print("\nCode Generation Tasks (Procedural Simulation)\n" + "="*50)
for task in code_gen.tasks:
    print(f"\n{task['name']}:")
    for steps in [1, 4, 8]:
        result = code_gen.evaluate(task['id'], steps, model_skill=0.75)
        status = '✓' if result['correct'] else '✗'
        print(f"  {steps} steps: {status} Tests: {result['tests_passed']}/10, Efficiency: {result['efficiency_score']:.2f}")
```

## Application 4: Domain-Specific Optimization

Design custom reward functions for your own tasks.

### NOTE: Reward Function Patterns
These reward function classes demonstrate different patterns for scoring performance. They take simulated attempt data and compute rewards. In production, the `attempt` dictionaries would come from actual model evaluations, but the reward function patterns shown here are real and directly applicable.

```python
class CustomRewardFunction:
    """Template for creating domain-specific reward functions."""
    
    def __call__(self, attempt: Dict) -> float:
        """Compute reward for an attempt.
        
        Args:
            attempt: dict with 'correctness', 'reasoning_length', etc.
        
        Returns:
            Reward value (typically 0-1 or -1 to 1)
        """
        raise NotImplementedError

class BalancedMathReward(CustomRewardFunction):
    """Reward both correctness and efficiency."""
    
    def __call__(self, attempt: Dict) -> float:
        correctness = attempt.get('correct', 0)
        reasoning_length = attempt.get('reasoning_length', 1)
        
        # Base reward for correctness
        base = 1.0 if correctness else 0.0
        
        # Efficiency bonus (prefer shorter proofs)
        if correctness:
            efficiency = max(0, 1 - reasoning_length / 10) * 0.2
            return base + efficiency
        
        return base

class ScaledDifficultyReward(CustomRewardFunction):
    """Harder problems get higher rewards."""
    
    def __call__(self, attempt: Dict) -> float:
        correctness = attempt.get('correct', 0)
        difficulty = attempt.get('difficulty', 0.5)
        
        if correctness:
            # Harder problems are worth more
            return 0.5 + (difficulty * 0.5)
        
        return 0.0

class ProcessRewardModel(CustomRewardFunction):
    """Reward intermediate steps, not just final answer."""
    
    def __call__(self, attempt: Dict) -> float:
        correctness = attempt.get('correct', 0)
        reasoning_length = attempt.get('reasoning_length', 0)
        intermediate_scores = attempt.get('step_scores', [])
        
        # Final answer matters most
        final_reward = 0.7 if correctness else 0.0
        
        # Intermediate steps help even if final is wrong
        if intermediate_scores:
            process_reward = np.mean(intermediate_scores) * 0.3
            final_reward += process_reward
        
        return final_reward

# Test different reward functions
test_attempt = {
    'correct': True,
    'reasoning_length': 3,
    'difficulty': 0.7,
    'step_scores': [0.8, 0.9, 1.0]
}

print("\nCustom Reward Functions\n" + "="*50)
print(f"\nTest attempt: Correct={test_attempt['correct']}, Steps={test_attempt['reasoning_length']}")
print(f"\nReward Function Results:")
print(f"  Balanced Math:        {BalancedMathReward()(test_attempt):.3f}")
print(f"  Scaled Difficulty:    {ScaledDifficultyReward()(test_attempt):.3f}")
print(f"  Process Reward Model:  {ProcessRewardModel()(test_attempt):.3f}")
```

## Application 5: Multi-Domain Evaluation

Compare performance across different reasoning tasks.

### NOTE: Procedural Simulation
This evaluation combines all previous applications (all procedurally simulated). The graphs show how reasoning depth theoretically affects performance across domains. In production, these would be actual results from training a real model with GRPO.

```python
# Evaluate performance across domains
def evaluate_all_domains(reasoning_steps: int = 5) -> Dict:
    """Evaluate model performance across all domains (ALL SIMULATED)."""
    
    results = {}
    
    # Math
    math_solver = MathProblemSolver()
    math_correct = sum([
        math_solver.evaluate(i, reasoning_steps)['correct']
        for i in range(len(math_solver.problems))
        for _ in range(5)
    ]) / (len(math_solver.problems) * 5)
    results['Math'] = math_correct
    
    # Logic
    logic_solver = LogicPuzzleSolver()
    logic_correct = sum([
        logic_solver.evaluate(i, reasoning_steps)
        for i in range(len(logic_solver.puzzles))
        for _ in range(5)
    ]) / (len(logic_solver.puzzles) * 5)
    results['Logic'] = logic_correct
    
    # Code
    code_gen = CodeGenerationTask()
    code_correct = sum([
        code_gen.evaluate(i, reasoning_steps)['correct']
        for i in range(len(code_gen.tasks))
        for _ in range(5)
    ]) / (len(code_gen.tasks) * 5)
    results['Code'] = code_correct
    
    return results

# Compare different reasoning depths
depths = [1, 3, 5, 8]
results_by_depth = {}

print("\nMulti-Domain Evaluation (Procedural Simulation)\n" + "="*50)

for depth in depths:
    results_by_depth[depth] = evaluate_all_domains(depth)
    print(f"\nWith {depth} reasoning steps:")
    for domain, accuracy in results_by_depth[depth].items():
        print(f"  {domain:<10}: {accuracy:.1%}")

# Visualization
fig, ax = plt.subplots(figsize=(10, 6))

for domain in ['Math', 'Logic', 'Code']:
    accuracies = [results_by_depth[d][domain] for d in depths]
    ax.plot(depths, accuracies, marker='o', label=domain, linewidth=2, markersize=8)

ax.set_xlabel('Reasoning Steps')
ax.set_ylabel('Accuracy')
ax.set_title('Multi-Domain Performance vs Reasoning Depth (Simulated)')
ax.legend(fontsize=11)
ax.grid(alpha=0.3)
ax.set_ylim([0, 1])

plt.tight_layout()
plt.savefig('phase3_multidomain.png', dpi=100, bbox_inches='tight')
plt.show()

print("\n✅ Multi-domain evaluation complete (all results are simulated)!")
```
