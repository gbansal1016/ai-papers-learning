## The Core Idea

Reinforcement Learning is learning through **trial and error** with **rewards**.

Think of training a dog:
- Dog tries an action (sits, jumps, runs)
- If good: give treat (reward) → dog learns to do more
- If bad: no treat → dog learns to do less

Eventually: dog learns actions that lead to treats!

## Example 1: Simple Reward

**Scenario:** A robot on a grid trying to reach a goal

```
Grid world:
S . . .      S = Start
. . . G      G = Goal
. . . .
. . . .

Robot can go: UP, DOWN, LEFT, RIGHT

Reward:
  - Reach goal: +10
  - Each step: -1 (cost to move)
  - Hit wall: -1 (can't move there)
```

**What the robot learns:**
- Moving randomly: lots of steps, low reward
- Moving toward goal: fewer steps, higher reward
- Eventually: optimal path to goal

**Key insight:** Rewards guide learning without explicit instruction!

## The RL Loop

```
┌─────────────────────────────────────────┐
│          The RL Loop                    │
│                                         │
│  1. Agent observes STATE               │
│      (current situation)                │
│            ↓                            │
│  2. Agent takes ACTION                 │
│      (based on policy/strategy)         │
│            ↓                            │
│  3. Environment gives REWARD            │
│      (feedback on action quality)       │
│            ↓                            │
│  4. Agent LEARNS                        │
│      (update policy to get more reward) │
│            ↓                            │
│  REPEAT (back to step 1)               │
│                                         │
└─────────────────────────────────────────┘
```

This repeats thousands or millions of times until the agent gets good!

## Key Concepts

### 1. STATE (s)
**What the agent observes about the world**

Examples:
- Robot's position on grid
- Game board configuration  
- Stock price history
- Robot sensor readings

**Question the agent asks:** "Where am I? What's the situation?"

---

### 2. ACTION (a)
**What the agent can do**

Examples:
- Move up, down, left, right
- Press button A or B
- Buy stock or sell
- Joint movements in robot

**Question:** "What should I do now?"

---

### 3. REWARD (r)
**Feedback from environment about action quality**

Examples:
- +1 for reaching goal
- -1 for bumping into wall
- +100 for winning game
- -10 for losing money

**Important:** Reward is usually delayed!
- In chess: you only get +1 for winning at the end
- But many moves lead to that win
- How do we know which moves helped?

**This is the credit assignment problem!**

---

### 4. POLICY (π)
**The agent's decision-making function**

Policy = mapping from state → action

```
π(state) = what action to take
```

Examples:
- "If goal is to the right, move right"
- "If opponent has 3 in a row, block"
- "If price is high, sell"

**RL's goal:** Learn a good policy!

---

### 5. RETURN (G)
**Total reward accumulated**

```
G = r₁ + r₂ + r₃ + ...
```

But with a trick: discount future rewards less

```
G = r₁ + 0.99*r₂ + 0.99²*r₃ + ...
```

**Why?** Immediate rewards are better than distant ones
- Like: \$100 now vs \$100 next year

## RL vs Supervised Learning

### Supervised Learning (Classification/Regression)
```
Input:  Image → What digit is this?
Label:  3
Loss:   Compare prediction to label
Learn:  Update weights to match labels

Key: We're given the RIGHT ANSWER
```

### Reinforcement Learning
```
State:  Position on grid
Action: Move right
Reward: -1 (took a step)
Learn:  Was this good? Better than alternatives?

Key: We're NOT given the right answer,
     just feedback on how good this was
```

**Analogy:**
- Supervised: Parent shows you the right spelling, you memorize
- RL: Parent says "wrong" or "right", you figure out the pattern

**RL is harder because:**
1. Rewards are delayed (was it this move or a previous move?)
2. No explicit "correct answer" (just feedback)
3. Exploration is hard (try bad actions to learn)
4. High variance (randomness matters)

## Exploration vs Exploitation

**Exploration:** Try new things to learn
- "Let me try moving in different directions"
- "Let me try different strategies"
- Necessary to discover good actions

**Exploitation:** Use what you know works
- "I know moving right gets me closer to goal"
- "Let me keep doing that"
- Necessary to get good rewards

**The Exploration-Exploitation Trade-off:**

Too much exploration:
- Try random actions forever
- Never actually achieve goal
- Wastes time

Too much exploitation:
- Stick with first decent action
- Never find better actions
- Stuck in local optima

**Ideal:** Balance exploration and exploitation
- Early: explore more (learn the environment)
- Late: exploit more (use knowledge)

**Real example:** Restaurant choice
- Exploration: Try new restaurant (might be great or terrible)
- Exploitation: Go to favorite restaurant (know it's good)
- Ideal: Try new restaurants sometimes, but mostly go to favorite

## Example 2: Multi-Armed Bandit

**Scenario:** Slot machines with unknown payouts

```
3 slot machines:
  Arm 1: 50% chance of \$1
  Arm 2: 80% chance of \$1  ← Best but you don't know it!
  Arm 3: 30% chance of \$1

You have 100 pulls total.
Goal: Maximize total money
```

**Naive approach: Equal exploration**
- Pull each arm 33 times to learn probabilities
- Then pull best arm 1 time
- Terrible! You spent most pulls learning

**Better approach: Explore-exploit balance**
- Pull each arm 5 times (20 pulls total)
- See which looks best (Arm 2: 4/5 = 80%)
- Pull Arm 2 for remaining 80 pulls
- Earn \$64 from best arm

**Key insight:** Need to balance learning and earning!

## Why RL Matters

### Problems RL Solves

**1. When we don't have labeled data**
- No one can label "best chess move"
- We just know who won
- RL learns from game outcome

**2. When multiple solutions exist**
- Many ways to reach a goal
- Reward guides agent to discover them
- Can find creative solutions

**3. When we want adaptive behavior**
- Robot adapts to environment
- Game AI adapts to player
- System adapts to new situations

**4. When we want efficient solutions**
- Shorter paths cost less
- Fewer moves to solve
- Optimal resource use

### Real-World Applications

- **AlphaGo:** Beat world champions at Go
- **Robotics:** Learn to walk, grasp, manipulate
- **Games:** Train agents to play Atari, Dota 2, StarCraft
- **Recommendation:** Netflix suggests movies
- **Autonomous vehicles:** Learn to drive
- **Finance:** Trade stocks/cryptocurrencies
- **Language models:** RLHF makes ChatGPT helpful
- **Control:** HVAC systems, traffic lights

## Simple Code Example

Let's see RL in the simplest possible way:

````python
import numpy as np

# Simple bandit problem
class SimpleBandit:
    """A simple slot machine with 3 arms"""
    
    def __init__(self):
        # True probabilities (agent doesn't know these)
        self.arm_probs = [0.2, 0.8, 0.4]  # Arm 1, 2, 3
    
    def pull(self, arm_id):
        """Pull an arm, get reward"""
        prob = self.arm_probs[arm_id]
        # 1 with probability prob, 0 otherwise
        return 1.0 if np.random.random() < prob else 0.0

# Create bandit
bandit = SimpleBandit()

# Simple learning: track average reward for each arm
num_pulls_per_arm = [0, 0, 0]
total_reward_per_arm = [0.0, 0.0, 0.0]

print("Learning Phase: Pull each arm 20 times")
print()

for arm_id in range(3):
    for _ in range(20):
        reward = bandit.pull(arm_id)
        num_pulls_per_arm[arm_id] += 1
        total_reward_per_arm[arm_id] += reward

# Calculate average reward for each arm
print("After exploring (20 pulls each):")
for arm_id in range(3):
    avg_reward = total_reward_per_arm[arm_id] / num_pulls_per_arm[arm_id]
    print(f"Arm {arm_id+1}: {avg_reward:.2%} average reward")

print(f"\nBest arm appears to be: Arm {np.argmax(total_reward_per_arm) + 1}")
print(f"(True best is Arm 2 with 80% probability)")
print()
print("Now we exploit: Keep pulling best arm")
print()

# Exploit best arm for remaining pulls
best_arm = np.argmax(total_reward_per_arm)
for _ in range(40):
    reward = bandit.pull(best_arm)
    num_pulls_per_arm[best_arm] += 1
    total_reward_per_arm[best_arm] += reward

print(f"Final stats (60 pulls total):")
for arm_id in range(3):
    avg = total_reward_per_arm[arm_id] / num_pulls_per_arm[arm_id]
    print(f"Arm {arm_id+1}: {total_reward_per_arm[arm_id]:.0f} total reward, {avg:.2%} average")

total_reward = sum(total_reward_per_arm)
print(f"\nTotal reward earned: {total_reward:.0f} out of 60 possible")
print(f"Success rate: {(total_reward/60):.1%}")
````

## What We Just Did

```
1. Agent explored (tried each arm 20 times)
   → Learned approximate value of each arm

2. Agent exploited (kept using best arm)
   → Earned rewards from what it learned

3. Result: Discovered best arm was #2
   → Total reward: ~48 out of 60
   → Would have earned more if perfect from start!
```

This is the essence of RL:
- **Learn** from rewards which actions are good
- **Act** on that knowledge to achieve goals
- **Trade-off** between learning and earning

## Key Takeaways

### Core RL Concepts
1. **State:** Current observation of environment
2. **Action:** What agent can do
3. **Reward:** Feedback on action quality
4. **Policy:** Mapping state → action
5. **Return:** Total accumulated reward

### Why RL is Different
- Learning from rewards, not labels
- Credit assignment problem (delayed rewards)
- Exploration-exploitation trade-off
- High variance (random)

### RL is Powerful Because
- No need for labeled data
- Can handle complex environments
- Discovers novel solutions
- Adapts to changes

### RL is Hard Because
- High variance (noisy learning signal)
- Sample inefficient (needs many trials)
- Credit assignment (which action caused reward?)
- Stability (small changes can break training)
